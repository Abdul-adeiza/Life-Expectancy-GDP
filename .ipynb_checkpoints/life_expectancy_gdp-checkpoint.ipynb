# LIFE EXPECTANCY AND GDP ANALYSIS OF SIX NATIONS
## Import Libraries and Load Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
all_data = pd.read_csv('all_data.csv')
```

```python
print(all_data.head())
```

## Summary Analysis

```python
print(all_data.info())
```

****From the summary statistics above, **Zimbabwe** has a lower life expectancy average of **50.09375** <br> 
between 2000 and 2015 when compared to the other 5 countries.

## Relationship Between Life Expectancy and GDP

```python
plt.figure(figsize=(15, 10))

plt.subplot(2, 3, 1)
sns.scatterplot(x = all_data['Life expectancy at birth (years)'][all_data.Country == 'Chile'], y = all_data.GDP[all_data.Country == 'Chile'],
                data = all_data, color = 'green')
plt.title('Chile')
plt.ylabel('GDP in Trillions of U.S Dollars')

plt.subplot(2, 3, 2)
sns.scatterplot(x = all_data['Life expectancy at birth (years)'][all_data.Country == 'China'], y = all_data.GDP[all_data.Country == 'China'],
                data = all_data, color = 'blue')
plt.title('China')
plt.ylabel('GDP in Trillions of U.S Dollars')

plt.subplot(2, 3, 3)
sns.scatterplot(x = all_data['Life expectancy at birth (years)'][all_data.Country == 'Germany'], y = all_data.GDP[all_data.Country == 'Germany'],
                data = all_data, color = 'red')
plt.title('Germany')
plt.ylabel('GDP in Trillions of U.S Dollars')

plt.subplot(2, 3, 4)
sns.scatterplot(x = all_data['Life expectancy at birth (years)'][all_data.Country == 'Mexico'], y = all_data.GDP[all_data.Country == 'Mexico'],
                data = all_data, color = 'yellow')
plt.title('Mexico')
plt.ylabel('GDP in Trillions of U.S Dollars')

plt.subplots_adjust(top = 1.2)

plt.subplot(2, 3, 5)
sns.scatterplot(x = all_data['Life expectancy at birth (years)'][all_data.Country == 'United States of America'], 
                y = all_data.GDP[all_data.Country == 'United States of America'], data = all_data, color = 'purple')
plt.title('United States of America')
plt.ylabel('GDP in Trillions of U.S Dollars')

plt.subplot(2, 3, 6)
sns.scatterplot(x = all_data['Life expectancy at birth (years)'][all_data.Country == 'Zimbabwe'], y = all_data.GDP[all_data.Country == 'Zimbabwe'],
                data = all_data, color = 'orange')
plt.title('Zimbabwe')
plt.ylabel('GDP in Trillions of U.S Dollars')
plt.savefig('GDP_VS_Life_Expectancy.jpg')

plt.show()
```

#### OBSERVATION
A. From the figure shown above;
1. In Chile, GDP and life expectancy can be seen to have an upward trend but GDP shows a decline at the 80.0 life expectancy mark.
2. In China and United State of America, GDP and life expectancy experienced an upward trend. As GDP rises, life expectancy rises.
3. In Germany and Mexico, GDP and life expectancy experienced and upward trend but nose dived at the 81.0 and 76.5 life expectancy mark.
4. In Zimbabwe, life expectancy started with a decline as GDP rises but took on a upward trend at the 0.6 GDP mark.

## Life Expectancy Over Time

```python
plt.figure(figsize=(15, 10))
plt.subplot(2, 3, 1)
plt.plot(all_data['Life expectancy at birth (years)'][all_data.Country == 'Chile'], all_data.Year[all_data.Country == 'Chile'], 
         color = 'green', marker = 's')
plt.title('Chile\'s Life Expectancy over time')
plt.xlabel('Life Expectancy')
plt.ylabel('Year (2000 - 2015)')

plt.subplot(2, 3, 2)
plt.plot(all_data['Life expectancy at birth (years)'][all_data.Country == 'China'], all_data.Year[all_data.Country == 'China'], 
         color = 'red', marker = 's')
plt.title('China\'s Life Expectancy over time')
plt.xlabel('Life Expectancy')
plt.ylabel('Year (2000 - 2015)')

plt.subplot(2, 3, 3)
plt.plot(all_data['Life expectancy at birth (years)'][all_data.Country == 'Germany'], all_data.Year[all_data.Country == 'Germany'], 
         color = 'blue', marker = 's')
plt.title('Germany\'s Life Expectancy over time')
plt.xlabel('Life Expectancy')
plt.ylabel('Year (2000 - 2015)')

plt.subplots_adjust(top = 1.2)

plt.subplot(2, 3, 4)
plt.plot(all_data['Life expectancy at birth (years)'][all_data.Country == 'Mexico'], all_data.Year[all_data.Country == 'Mexico'], 
         color = 'Yellow', marker = 's')
plt.title('Mexico\'s Life Expectancy over time')
plt.xlabel('Life Expectancy')
plt.ylabel('Year (2000 - 2015)')

plt.subplot(2, 3, 5)
plt.plot(all_data['Life expectancy at birth (years)'][all_data.Country == 'United States of America'], 
         all_data.Year[all_data.Country == 'United States of America'], 
         color = 'Brown', marker = 's')
plt.title('USA\'s Life Expectancy over time')
plt.xlabel('Life Expectancy')
plt.ylabel('Year (2000 - 2015)')

plt.subplot(2, 3, 6)
plt.plot(all_data['Life expectancy at birth (years)'][all_data.Country == 'Zimbabwe'], all_data.Year[all_data.Country == 'Zimbabwe'], 
         color = 'Purple', marker = 's')
plt.title('Zimbabwe\'s Life Expectancy over time')
plt.xlabel('Life Expectancy')
plt.ylabel('Year (2000 - 2015)')

plt.show()
```

#### OBSERVATION
1. China, Germany, United States of America expreienced a constant life expectancy increase over time.
2. Zimbabwe's life expectancy started with a decline from the year 2000 to 2004 but showed an upward trend between 2004 to 2015.
3. Chile's and Mexico's life expectancy had an overall increase over time but experienced a decrease between 2008 to 2010 and 2006 to 2008.

## GDP Over Time

```python
plt.figure(figsize=(15, 10))
plt.subplot(2, 3, 1)
plt.plot(all_data['GDP'][all_data.Country == 'Chile'], all_data.Year[all_data.Country == 'Chile'], 
         color = 'green', linestyle = ':', marker = 's')
plt.title('Chile\'s GDP over time')
plt.xlabel('GDP in Trillions of U.S Dollars')
plt.ylabel('Year (2000 - 2015)')

plt.subplot(2, 3, 2)
plt.plot(all_data['GDP'][all_data.Country == 'China'], all_data.Year[all_data.Country == 'China'], 
         color = 'red', linestyle = ':', marker = 's')
plt.title('China\'s GDP over time')
plt.xlabel('GDP in Trillions of U.S Dollars')
plt.ylabel('Year (2000 - 2015)')

plt.subplot(2, 3, 3)
plt.plot(all_data['GDP'][all_data.Country == 'Germany'], all_data.Year[all_data.Country == 'Germany'], 
         color = 'blue', linestyle = ':', marker = 's')
plt.title('Germany\'s GDP over time')
plt.xlabel('GDP in Trillions of U.S Dollars')
plt.ylabel('Year (2000 - 2015)')

plt.subplots_adjust(top = 1.2)

plt.subplot(2, 3, 4)
plt.plot(all_data['GDP'][all_data.Country == 'Mexico'], all_data.Year[all_data.Country == 'Mexico'], 
         color = 'Yellow', linestyle = ':', marker = 's')
plt.title('Mexico\'s GDP over time')
plt.xlabel('GDP in Trillions of U.S Dollars')
plt.ylabel('Year (2000 - 2015)')

plt.subplot(2, 3, 5)
plt.plot(all_data['GDP'][all_data.Country == 'United States of America'], 
         all_data.Year[all_data.Country == 'United States of America'], 
         color = 'Brown', linestyle = ':', marker = 's')
plt.title('USA\'s GDP over time')
plt.xlabel('GDP in Trillions of U.S Dollars')
plt.ylabel('Year (2000 - 2015)')

plt.subplot(2, 3, 6)
plt.plot(all_data['GDP'][all_data.Country == 'Zimbabwe'], all_data.Year[all_data.Country == 'Zimbabwe'], 
         color = 'Purple', linestyle = ':', marker = 's')
plt.title('Zimbabwe\'s GDP over time')
plt.xlabel('GDP in Trillions of U.S Dollars')
plt.ylabel('Year (2000 - 2015)')

plt.show()
```

#### OBSERVATION
1. China and United States of America GDP increased overtime.
2. Chile, Germany and Mexico showed an overall increase in GDP over time. <br> Although, the three nations started showing a decrease between
   the year 2012 1nd 2014.
3. Zimbabwe's GDP started off with a decline from 2000 to 2008. From 2008, the nation adopted an upward trend. 

## CONCLUSION
It was observed that the GDP and Life Expectancy of all six nations had an overall growth over time.<br> 
Zimbabwe had the lowest average life expectancy of **50.09375** while the other 5 nations had an average
